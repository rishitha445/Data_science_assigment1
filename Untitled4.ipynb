```python
import pandas as pd

# Loading the CSV file to a DataFrame
df = pd.read_csv('social_media_usage.csv')

# Display the first rows of a dataframe
df.head()
```

```python
import matplotlib.pyplot as plt
def plot_avg_posts_per_platform(data):
    """
    Creating a bar chart showing the average posts per day on each social media platform.
    """
    avg_posts = df.groupby('App')['Posts_Per_Day'].mean().sort_values()
    plt.figure(figsize=(10, 6))
    # plotting the bar chart graph
    avg_posts.plot(kind='bar', color='purple')
    plt.title("Average Posts Per Day by Platform")
    plt.xlabel("Social Media Platform")
    plt.ylabel("Average Posts Per Day")
    plt.show()

# plotting the graphs
plot_avg_posts_per_platform(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 
def plot_minutes_vs_follows(data):
    """
    Creates a scatter plot showing the relationship between daily minutes spent and follows per day.
    """
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='Daily_Minutes_Spent', y='Follows_Per_Day',  color='red', data=data, alpha=0.6,)
    plt.xlabel("Daily Minutes Spent")
    plt.ylabel("Follows Per Day")
    plt.title("Relationship between Daily Minutes Spent and Follows Per Day")
    plt.show()

# plotting the graphs
plot_minutes_vs_follows(df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_correlation_heatmap(data):
    """
    Creates a heatmap to visualize the correlation between daily minutes spent, posts per day, likes per day, 
    and follows per day.
    """
    # Calculate correlation matrix
    correlation = data[['Daily_Minutes_Spent', 'Posts_Per_Day', 'Likes_Per_Day', 'Follows_Per_Day']].corr()
    
    # Plot
    plt.figure(figsize=(6, 6))
    sns.heatmap(correlation, annot=True, cmap='cividis', fmt=".2f", linewidths=0.7)
    plt.title("Correlation Heatmap of User Activity Metrics")
    plt.show()

# Generating the heatmap
plot_correlation_heatmap(df)
```
